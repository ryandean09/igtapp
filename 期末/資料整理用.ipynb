```python
"""檢查檔案
"""

import os


def check_folders_for_same_files(path):
    files_sets = []  # 儲存每個資料夾中的檔案集合

    # 列出指定路徑下的所有檔案與資料夾
    for root, dirs, files in os.walk(path):
        # 只處理資料夾而非子資料夾
        if root == path:
            for folder in dirs:
                folder_path = os.path.join(root, folder)
                files_set = set()  # 建立空的檔案集合
                for file in os.listdir(folder_path):
                    files_set.add(file)  # 將檔案加入集合
                files_sets.append(files_set)  # 將集合加入列表

    # 檢查所有資料夾中的檔案集合是否都相同
    return all(files_set == files_sets[0] for files_set in files_sets)


# 指定要檢查的路徑
folder_path = r"C:\Users\D\Desktop\0228\人口"

# 獲取結果
result = check_folders_for_same_files(folder_path)

# 輸出結果
if result:
    print("每個資料夾中都含有相同的檔案。")
else:
    print("資料夾中含有不同的檔案。")
```

```python
def read_file_with_encoding(file_path):
    import pandas as pd

    # 定義可能的編碼方式列表
    encodings = ["utf-8", "big5", "gbk"]
    for encoding in encodings:
        try:
            df = pd.read_csv(file_path, encoding=encoding)
            return df, encoding  # 如果成功讀取，則返回DataFrame和使用的編碼方式
        except UnicodeDecodeError:
            pass  # 如果出現解碼錯誤，繼續嘗試下一個編碼方式
    # return None, None  # 如果沒有成功讀取，則返回None和None
    return None, None
```

```python
import pandas as pd
import os


def save_dataframe_csv_excel(df, filename):
    """
    將 DataFrame 存儲為 CSV 和 Excel 檔案，並打開 Excel 檔案。

    Args:
        df (pandas.DataFrame): 要存儲的 DataFrame。
        filename (str): 要保存的文件名，不包含檔案擴展名。
    """
    # 將 DataFrame 存儲為 CSV 檔案
    csv_filename = filename + ".csv"
    df.to_csv(csv_filename, index=False, encoding="utf-8")
    print(f"已保存為 CSV 檔案：{csv_filename}")

    # 將 DataFrame 存儲為 Excel 檔案
    excel_filename = filename + ".xlsx"
    df.to_excel(excel_filename, index=False)
    print(f"已保存為 Excel 檔案：{excel_filename}")

    # 打開 Excel 檔案
    os.startfile(excel_filename)
    print("已打開 Excel 檔案")


def save_dataframe_excel_and_open(df, filename):
    """
    將 DataFrame 存儲為Excel 檔案，並打開 Excel 檔案。

    Args:
        df (pandas.DataFrame): 要存儲的 DataFrame。
        filename (str): 要保存的文件名，不包含檔案擴展名。
    """

    # 將 DataFrame 存儲為 Excel 檔案
    excel_filename = filename + ".xlsx"
    df.to_excel(excel_filename, index=False)
    print(f"已保存為 Excel 檔案：{excel_filename}")

    # 打開 Excel 檔案
    os.startfile(excel_filename)
    print("已打開 Excel 檔案")


# 使用範例
# save_dataframe_csv_excel(變數名稱, "輸出檔案名")
```

```python
import pandas as pd


def analyze_nan_and_zero_values(
    data, threshold, sort_column="零值+NAN佔比 (%)", ascending=False
):
    """
    檢查0值和 nan 的數量
    threshold 為 百分比，例threshold=20，會回傳 零值+NAN佔比 (%) 大於20% 的
    當 threshold = 0 會回傳全部
    """
    # 如果輸入是文件路徑，讀取CSV文件並轉換成Pandas DataFrame
    if isinstance(data, str):
        file_source_pd = pd.read_csv(data)
    # 如果輸入是Pandas DataFrame，直接使用它
    elif isinstance(data, pd.DataFrame):
        file_source_pd = data
    else:
        raise ValueError(
            "Invalid input type. Input must be either file path or Pandas DataFrame."
        )

    # 計算這個範圍內每一欄的空白值(NaN)數量
    nan_values_per_column_in_range = file_source_pd.isnull().sum(axis=0)

    # 計算每個欄位的總數據量
    total_data_per_column = len(file_source_pd)

    # 計算每個欄位的空白值(NaN)佔比
    nan_percentage_per_column = (
        nan_values_per_column_in_range / total_data_per_column
    ) * 100

    # 計算這個範圍內每一欄的零值數量
    zero_values_per_column_in_range = (file_source_pd == 0).sum(axis=0)

    # 計算每個欄位的零值佔比
    zero_percentage_per_column = (
        zero_values_per_column_in_range / total_data_per_column
    ) * 100

    # 計算每個欄位的零值和NaN值的總數
    total_zero_and_nan_per_column = (
        zero_values_per_column_in_range + nan_values_per_column_in_range
    )

    # 計算每個欄位的零值和NaN值總數的佔比
    total_zero_and_nan_percentage_per_column = (
        total_zero_and_nan_per_column / total_data_per_column
    ) * 100

    # 將結果轉換為 DataFrame
    values_df = pd.DataFrame(
        {
            "欄位名稱": nan_values_per_column_in_range.index,
            "零值+NAN佔比 (%)": total_zero_and_nan_percentage_per_column.values.round(
                2
            ),
            "空白值(NaN)數量": nan_values_per_column_in_range.values,
            "空白值(NaN)佔比 (%)": nan_percentage_per_column.values.round(2),
            "零值數量": zero_values_per_column_in_range.values,
            "零值佔比 (%)": zero_percentage_per_column.values.round(2),
        }
    )

    # 將 DataFrame 存儲到 CSV 檔案中
    values_df.to_csv("零值與空白值統計.csv", index=False, encoding="utf-8-sig")

    # 如果 threshold 等於 0，則返回所有欄位
    if threshold == 0:
        return values_df

    # 篩選出 '零值+NAN佔比 (%)' 大於指定閾值的欄位
    filtered_values_df = values_df[values_df["零值+NAN佔比 (%)"] > threshold]

    # 根據指定的列進行排序
    sorted_values_df = filtered_values_df.sort_values(
        by=sort_column, ascending=ascending
    )

    return sorted_values_df


# 測試
# result_df = analyze_nan_and_zero_values(clean_data, threshold=0, sort_column='欄位名稱', ascending=False)
# print(result_df)
```

# 填補營業時間中位數

```python
""" 填補營業時間中位數
"""

import pandas as pd

filled_source_path = r"..\飲料店總表0307final01_補上人氣_補值_hg.xlsx"

file_data_pd = pd.read_excel(filled_source_path)


# 篩選出不包含空白值的資料
clean_data = file_data_pd.dropna(
    subset=[
        "Monday_open_hours",
        "Tuesday_open_hours",
        "wendesday_open_hours",
        "Thursday_open_hours",
        "Friday_open_hours",
    ]
)

# 按照district和neighborhood分組，並計算每組的Monday_open_hours中位數
district_neighborhood_median_hours = (
    clean_data.groupby(["district", "neighborhood"])[
        [
            "Monday_open_hours",
            "Tuesday_open_hours",
            "wendesday_open_hours",
            "Thursday_open_hours",
            "Friday_open_hours",
        ]
    ]
    .median()
    .reset_index()
)

district_median_hours = (
    clean_data.groupby(["district"])[
        [
            "Monday_open_hours",
            "Tuesday_open_hours",
            "wendesday_open_hours",
            "Thursday_open_hours",
            "Friday_open_hours",
        ]
    ]
    .median()
    .reset_index()
)


# 將中位數資料合併回原始資料中
fill_neighborhood_workday_open_hours_pd = pd.merge(
    file_data_pd,
    district_neighborhood_median_hours,
    on=["district", "neighborhood"],
    how="left",
    suffixes=('', '_neighborhood_median')
)

# 將中位數資料合併回原始資料中
fill_district_workday_open_hours_pd = pd.merge(
    fill_neighborhood_workday_open_hours_pd,
    district_median_hours,
    on=["district"],
    how="left",
    suffixes=('', '_district_median')
)


# # 儲存到Excel檔案
# filename = "三峽龍學理.xlsx"  # Excel檔案名稱
save_dataframe_csv_excel(fill_district_workday_open_hours_pd, "區域中位數填補資料")
```
